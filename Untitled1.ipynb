```python
from nltk.corpus import wordnet as wn
```

```python
with open("./words.txt") as f:
    words = f.readlines()

words = [w.strip() for w in words] 
```

```python
master_list = []
for w in words:
    for s in wn.synsets(w):
        master_list.extend(wn.synset(s.name()).lemma_names())
        for x in wn.synset(s.name()).hyponyms():
            master_list.extend(wn.synset(x.name()).lemma_names())
        for y in wn.synset(s.name()).hypernyms():
            master_list.extend(wn.synset(y.name()).lemma_names())
        for a in wn.synset(s.name()).part_holonyms():
            master_list.extend(wn.synset(a.name()).lemma_names())
        for b in wn.synset(s.name()).part_meronyms():
            master_list.extend(wn.synset(b.name()).lemma_names())
        for c in wn.synset(s.name()).entailments():
            master_list.extend(wn.synset(c.name()).lemma_names())

l = set()
for w in master_list:
    if not (w in l or "_" in w or "-" in w or w.isupper()):
        l.add(w)

print(l)
```
